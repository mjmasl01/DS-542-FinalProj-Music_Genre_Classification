# CNN'ing data

```python
import librosa

genres_path = "/Users/matthewmaslow/Desktop/DS-542 - Deep Learning/Final Project/data/genres_original"

features_list = []

for genre in os.listdir(genres_path):
    genre_path = os.path.join(genres_path, genre)
    if os.path.isdir(genre_path):
        for file in os.listdir(genre_path):
            if file.endswith('.wav'):
                file_path = os.path.join(genre_path, file)
                try:
                    y, sr = librosa.load(file_path, duration=30)
                    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).mean(axis=1)
                    chroma = librosa.feature.chroma_stft(y=y, sr=sr).mean(axis=1)
                    spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr).mean(axis=1)
                    features = {
                        "filename": file,
                        "label": genre,
                        **{f"mfcc{i+1}": mfcc for i, mfcc in enumerate(mfccs)},
                        **{f"chroma{i+1}": chroma_val for i, chroma_val in enumerate(chroma)},
                        **{f"spectral_contrast{i+1}": contrast for i, contrast in enumerate(spectral_contrast)},
                    }
                    features_list.append(features)
                except Exception as e:
                    print(f"Error processing file {file}: {e}")

features_df = pd.DataFrame(features_list)
features_df.to_csv("/Users/matthewmaslow/Desktop/DS-542 - Deep Learning/Final Project/data/audio_features.csv", index=False)
print("Audio features saved to 'audio_features.csv'")
```

```python
import librosa.display
import matplotlib.pyplot as plt

genres_path = "/Users/matthewmaslow/Desktop/DS-542 - Deep Learning/Final Project/data/genres_original"
spectrograms_path = "/Users/matthewmaslow/Desktop/DS-542 - Deep Learning/Final Project/data/spectrograms"
os.makedirs(spectrograms_path, exist_ok=True)

for genre in os.listdir(genres_path):
    genre_path = os.path.join(genres_path, genre)
    genre_spectrogram_path = os.path.join(spectrograms_path, genre)
    os.makedirs(genre_spectrogram_path, exist_ok=True)

    if os.path.isdir(genre_path):
        for file in os.listdir(genre_path):
            if file.endswith('.wav'): 
                file_path = os.path.join(genre_path, file)
                try:
                    y, sr = librosa.load(file_path, duration=30)
                    plt.figure(figsize=(10, 4))
                    S = librosa.feature.melspectrogram(y=y, sr=sr)
                    S_dB = librosa.power_to_db(S, ref=np.max)
                    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
                    plt.colorbar(format='%+2.0f dB')
                    plt.title(f"Spectrogram of {file}")
                    plt.tight_layout()
                    save_path = os.path.join(genre_spectrogram_path, f"{file[:-4]}.png")
                    plt.savefig(save_path)
                    plt.close()
                except Exception as e:
                    print(f"Error generating spectrogram for {file}: {e}")

print("Spectrograms saved to the 'spectrograms' directory.")
```

```python
spectrograms_path = "/Users/matthewmaslow/Desktop/DS-542 - Deep Learning/Final Project/data/spectrograms"

# Remove .DS_Store and other non-genre items
for item in os.listdir(spectrograms_path):
    item_path = os.path.join(spectrograms_path, item)
    if not os.path.isdir(item_path):
        print(f"Removing: {item}")
        os.remove(item_path)
```

```python
datagen = ImageDataGenerator(validation_split=0.2, rescale=1./255)

train_data = datagen.flow_from_directory(
    spectrograms_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    spectrograms_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

print("Class indices after cleanup:", train_data.class_indices)
print("Number of classes after cleanup:", len(train_data.class_indices))
```

```python
from tensorflow.keras import layers, models, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Verify the number of classes dynamically
num_classes = len(train_data.class_indices)

# Define the CNN architecture
model = models.Sequential([
    Input(shape=(128, 128, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')  # Dynamically match the number of classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(
    train_data,
    epochs=10,
    validation_data=val_data
)
```

## Plots

```python
import matplotlib.pyplot as plt

# Plot accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
val_loss, val_accuracy = model.evaluate(val_data)
print(f"Validation Loss: {val_loss}")
print(f"Validation Accuracy: {val_accuracy}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

file_path = '/Users/matthewmaslow/Desktop/DS-542 - Deep Learning/Final Project/data/audio_features.csv'
features_df = pd.read_csv(file_path)

X = features_df.drop(columns=['filename', 'label']).values
y = features_df['label']

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

X_train, X_val, y_train, y_val = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 15, 21, 25],
    'metric': ['euclidean', 'manhattan', 'minkowski', 'chebyshev', 'hamming', 'canberra', 'braycurtis'],
    'weights': ['uniform', 'distance']
}

grid_search = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best accuracy during cross-validation:", grid_search.best_score_)

best_knn = grid_search.best_estimator_
y_val_pred = best_knn.predict(X_val)

print(classification_report(y_val, y_val_pred, target_names=label_encoder.classes_))
```
